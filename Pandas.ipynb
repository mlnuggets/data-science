# Pandas Series

### Create a Series

```python
sample_list_to_series = pd.Series([300, 240, 160]) # pass in the python list into Series method
print('List to Series:')
print(sample_list_to_series)

sample_ndarray_to_series = pd.Series(np.array([90, 140, 80])) # pass the numpy array in the Series method
print('Numpy array to Series:')
sample_ndarray_to_series
```

### Access value with index

```python
print(sample_list_to_series[1])
print(sample_ndarray_to_series[2])
```

### Labeling series data with `index`

```python
sample_list_to_series = pd.Series([300, 240, 160], index = ['Ninja_HP', 'BMW_HP', 'Damon_HP'])
print(sample_list_to_series)

sample_ndarray_to_series = pd.Series(np.array([90, 140, 80]), index = ['valx', 'valy', 'valz'])
print(sample_ndarray_to_series)
```

### Retreive Data with the index lable values

```python
print(sample_list_to_series['BMW_HP'])
print(sample_ndarray_to_series['valy'])
```

### Series from a dictionary

```python
sample_dict = {'Ninja_HP': 300, 'BMW_HP': 240, 'Damon_HP': 160}
print(pd.Series(sample_dict))
```

### Retrieve multiple values

```python
print(sample_list_to_series[['Ninja_HP', 'Damon_HP']])
```

# Column selection, addition, deletion

### Column Selection

```python
dict_of_lists = {'Model' : ['Bentley', 'Toyota', 'Audi', 'Ford'], 'Weight' : [1400.8, 2500, 1600, 1700]}
indexed_data_frame = pd.DataFrame(dict_of_lists, index = ['model_1', 'model_2', 'model_3', 'model_4'])
print(indexed_data_frame['Weight'])
```

### Adding Column to the DataFrame

```python
dict_of_series = {'HP': pd.Series([300, 240, 160], index = ['Ninja', 'BMW', 'Damon']), 'speed' : pd.Series(np.array([280, 300, 260, 200]), index = ['Ninja', 'BMW', 'Damon', 'Suzuki'])}
bikes_data_df= pd.DataFrame(dict_of_series)
bikes_data_df['Fuel Consumption'] = pd.Series(np.array(['27Km/L', '24Km/L', '30Km/L', '22Km/L']), index = ['Ninja', 'BMW', 'Damon', 'Suzuki'])
print(bikes_data_df)
```

### Deleting a column with `pop` and `delete` functions

```python
# Delete or pop Speed and Fuel consumption columns from bikes data above
# using pop function
bikes_data_df.pop('speed')
print(bikes_data_df)

#using delete function
del bikes_data_df['Fuel Consumption']
print(bikes_data_df)
```

### Reading data from SQL database

```python
import sqlite3
conn = sqlite3.connect('vehicle_database') 
c = conn.cursor()
```

```python
# let's create a database and insert values with sqlite3
try:
  c.execute('''
            CREATE TABLE IF NOT EXISTS vehicle_data
            ([vehicle_id] INTEGER PRIMARY KEY, [vehicle_model] TEXT, [weight] INTEGER, [color] TEXT)
            ''')                     
  conn.commit()

# insert values into tables

  c.execute('''
            INSERT INTO vehicle_data (vehicle_id, vehicle_model, weight, color)

                  VALUES
                  (1,'Bentley',1400,'Blue'),
                  (2,'Toyota',2500,'Green'),
                  (3,'Audi',1600,'Black'),
                  (4,'Ford',1700,'White')
            ''')
  conn.commit()
except:
  print('DB exists!')
```

### Read Data from the sql database with `read_sql_query()`



```python
cars_df = pd.read_sql_query("SELECT * FROM vehicle_data", conn)
cars_df.set_index('vehicle_id')
```

# Cleaning data

## First let's shuffle the DataFrame

```python
shuffled_melbourne_data = melbourne_data.sample(frac=1, random_state=1) # produce te same shuffled data each time
shuffled_melbourne_data
```

### Dropping unecessary coumns

```python
# Define a list names of all the columns we want to drop
columns_to_drop = ['Method',
                   'SellerG',
                   'Date',
                   'Bathroom',
                   'Car',
                   'Landsize',
                   'BuildingArea',
                   'YearBuilt']
shuffled_melbourne_data.drop(columns = columns_to_drop, inplace=True, axis=1)
shuffled_melbourne_data
```

### Tranformations

*Merge method*

```python
price_means = shuffled_melbourne_data.groupby('No_ofBedrooms')['Price'].mean().rename('MeanPrice').reset_index()
df_1 = shuffled_melbourne_data.merge(price_means)
df_1
```

*`transform` operation*

```python
shuffled_melbourne_data['MeanPrice'] = shuffled_melbourne_data.groupby('No_ofBedrooms')['Price'].transform('mean')
shuffled_melbourne_data
```

# Pivot table

```python
pivot_tbl = shuffled_melbourne_data.pivot_table(index='HousingType')
pivot_tbl
```

### Pivot table for specific column `values=`

```python
pivot_tbl = shuffled_melbourne_data.pivot_table(index='HousingType', values=['Price', 'No_ofRooms'])
pivot_tbl
```

### `aggfunc=` aggregation methods

```python
pivot_tbl_sum = shuffled_melbourne_data.pivot_table(index='Regionname', aggfunc='sum', values='Propertycount')
pivot_tbl_sum
```

###  Multiple aggregation methods

```python
pivot_tbl_sum_max = shuffled_melbourne_data.pivot_table(index='Regionname', aggfunc=['sum', 'max'], values='Propertycount')
pivot_tbl_sum_max
```

### Different aggreation methods

```python
pivot_tbl = shuffled_melbourne_data.pivot_table(index='Regionname', values=['Price', 'Propertycount'], aggfunc={'Price':'sum', 'Propertycount':'mean'})
pivot_tbl
```

### Split data by column

```python
pivot_tbl = shuffled_melbourne_data.pivot_table(index='Regionname', values=['Price', 'Propertycount'], aggfunc={'Price':'sum', 'Propertycount':'mean'}, columns='HousingType')
pivot_tbl
```

### Adding totals `margin=`

```python
pivot_tbl = shuffled_melbourne_data.pivot_table(index='Regionname', values='Propertycount', margins=True, margins_name='Totals', aggfunc='sum', columns='HousingType')
pivot_tbl
```

```python
price_count = shuffled_melbourne_data.groupby('Address')['Price'].sum()
print(price_count)
```

```python
small_groups = price_count[price_count < 1500000]
large_groups = price_count[price_count > 1500000]

small_sums = pd.Series([small_groups.sum()], index=["Other"])
large_groups = large_groups.append(small_sums)

large_groups.iloc[0:10].plot(kind="pie", label="", title='Ratio prices at addresses')
plt.show()
```
