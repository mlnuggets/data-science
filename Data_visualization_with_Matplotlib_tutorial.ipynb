# Matplotlib Pyplot
Pyplot allows you to implicitly and automatically create figures and axes to achieve the desired plots.
* For creating plots quickly without first instantiaiting figures and the axes.
* If you have prior experience with MATLAB, then pyplot offers almost similar syntax.
* Pyplot is state-based as it modifies the same figure
*The figure is preserved through various function calls.

## Simple plot with pyplot

```python
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)
plt.plot(x, y)
plt.show()      
```

Providing a single list to the plot() function:

```python
arr_points= np.array([20, 30, 40, 50, 60])
plt.plot(arr_points)
```

### Lables and titles in Pyplot

`xlabel()`, `title()` and `ylabel()` functions set the labels for x and y axis

```python
x = np.linspace(50, 80, 20)
y = np.linspace(90, 300, 20)
z = np.linspace(150, 300, 20)
plt.plot(x, y)

plt.title('Relation of weight to height')
plt.xlabel('Heights in inches')
plt.ylabel('Weights in pounds')
```

### Customizing plotted lines


```python
x = np.linspace(50, 80, 20)
y = np.linspace(90, 300, 20)
z = np.linspace(150, 300, 20)
plt.plot(x, y, linestyle='-', color='tab:green', linewidth='15.4') # linestyle, color, linewidth:
# SAME AS
#plt.plot(x, y, ls='-.', c='y', lw='15.4')

plt.title('Relation of weight to height')
plt.xlabel('Heights in inches')
plt.ylabel('Weights in pounds')
```

## Building Subplots with `plt.subplots() function`

```python
x = np.linspace(0,4*np.pi,20)
y = np.sin(x)

plt.subplot(1, 2, 1) # first plot
plt.plot(x, y, ls='-.', c='tab:green', marker='*')
plt.ylabel('sin x')
plt.xlabel('x')

z = np.cos(x)

plt.subplot(1, 2, 2) # second plot
plt.plot(x, z, ls='-.', c='y', marker='^')
plt.ylabel('cos x')
plt.xlabel('x')

plt.tight_layout()
```

Example 2 : Includes more subplots and setting the figure size with `plt.figure()`

```python
X = np.linspace(0,4*np.pi,20)
ypoints1 = np.sin(X)

plt.figure(figsize= (10, 7))

plt.subplot(2, 2, 1) # plot 1
plt.plot(X, ypoints1, ls='--', c='tab:green')
plt.ylabel('sin x')
plt.xlabel('X values')

ypoints2 = np.cos(X)

plt.subplot(2, 2, 2) # plot 2
plt.plot(X, ypoints2, ls='-.', c='y')
plt.ylabel('cos x')
plt.xlabel('X values')

ypoints3 = np.square(X)

plt.subplot(2, 2, 3) # plot 3
plt.plot(X, ypoints3, ls='-', c='r')
plt.ylabel('X squared values')
plt.xlabel('X values')

ypoints4 = X**3

plt.subplot(2, 2, 4) # plot 4
plt.plot(X, ypoints4, ls=':', c=(0.3, 0.4, 0.7, 0.8))
plt.ylabel('X cubed values')
plt.xlabel('X values')

plt.suptitle('Multiple plots')

plt.tight_layout()
```

### Markers

```python
x = np.linspace(0,4*np.pi,20)
y = np.sin(x)
plt.plot(x, y, ls='-.', c='tab:green', marker='*') # star marker
plt.show()    
```

### Setting marker sizes and color
The marker size can be set with `markersize` or `ms` keyword.
`markeredgecolor` or `mec` keyword sets the edge color of the marker.
The color inside the marker is set with `markerfacecolor` or `mfc` keyword.

```python
x = np.linspace(0,4*np.pi,20)
ypoints1 = np.sin(x)

plt.figure(figsize= (10, 7))

plt.subplot(2, 2, 1)
plt.plot(x, ypoints1, ls='--', c='k', marker='o', ms='15')
plt.title('markersize(ms)')

ypoints2 = np.cos(x)

plt.subplot(2, 2, 2)
plt.plot(x, ypoints2, ls='-.', c='k', marker='o', ms='15', mec='r')
plt.title('markeredgecolor(mec)-red')

ypoints3 = np.arccos(ypoints2)

plt.subplot(2, 2, 3)
plt.plot(x, ypoints3, ls='-', c='k', marker='o', ms='15', mec='r', mfc='hotpink')
plt.title('markerfacecolor(mfc)-color inside marker-hotpink')
plt.tight_layout()
```

# Object Oriented Interface

### Creating a figure

```python
import math
x = np.arange(0, math.pi*5, 0.05)
y = np.sin(x)

fig = plt.figure(figsize=(8, 5), facecolor='y', edgecolor='tab:pink', linewidth=20) # creating the figure object

ax = fig.add_axes([0, 0, 1, 1]) # adding a axes

ax.plot(x, y)
```

### Adding Axes to figure
The Axes class makes it possible to create subplots on the figure.

```python
import math
# plot multiple axes in the same axes object

x = np.arange(0, math.pi*5, 0.05)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize=(8, 5)) # creating the figure object

ax = fig.add_axes([1, 1, 1, 1]) # adding an axes

ax1 = ax.plot(x, y, c='tab:green')
ax2 = ax.plot(x, z, c='tab:red')
```

## Adding Labels title and legend

```python
import math
x = np.arange(0, math.pi*5, 0.05)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize=(8, 5)) # creating the figure object

ax = fig.add_axes([1, 1, 1, 1]) # adding an axes

ax1 = ax.plot(x, y, c='tab:green')
ax2 = ax.plot(x, z, c='tab:red')

# setting the title of the figure
ax.set_title('sin and cos x')

# setting x and y labels
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

# setting Legends
ax.legend(['SIN X', 'COS X'], loc='upper right', bbox_to_anchor=(0.72, 1), fontsize='medium')
plt.show()
```

# Creating Multiple plots/ subplots

### Adding subplots with `add_axes()`

```python
months = np.array(['Jan', 'Feb', 'March', 'April', 'May', 'Jun'])
savings = np.array([8500, 14700, 12850, 13900, 23650, 15500])
expenses = np.array([1500, 1200, 1000, 500, 2500, 3000, 700])
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])


# Instantiate figure
fig = plt.figure(figsize=(5, 4), facecolor='pink', edgecolor='y', linewidth=5)
# Add first axes
ax1 = fig.add_axes([0, 0.1, 0.8, 0.8])
ax1.set_title('Monthly savings', fontsize=15)
ax1.set_xlabel('Months', fontsize=15)
ax1.set_ylabel('Savings', fontsize=15)
# Add second axes
ax2 = fig.add_axes([1, 0.1, 0.8, 0.8])

ax2.set_title('Average Day expenses', fontsize=15)
ax2.set_xlabel('Days', fontsize=15)
ax2.set_ylabel('expenses', fontsize=15)

# plot data
ax1.plot(months, savings, c='r')
ax2.plot(days, expenses, c='green')
```

Example two: full control of where we want Axes placed

```python
months = np.array(['Jan', 'Feb', 'March', 'April', 'May', 'Jun'])
savings = np.array([8500, 14700, 12850, 13900, 23650, 15500])
expenses = np.array([1500, 1200, 1000, 500, 2500, 3000, 700])
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])


# Instantiate figure
fig = plt.figure(figsize=(7, 4), facecolor='pink', edgecolor='y', linewidth=5)
# Add first axes
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.set_title('Monthly savings', fontsize=15)
ax1.set_xlabel('Months', fontsize=15)
ax1.set_ylabel('Savings', fontsize=15)
# Add second axes
ax2 = fig.add_axes([0.25, 0.5, 0.3, 0.3]) #inset Axes

ax2.set_title('Average Day expenses')
ax2.set_xlabel('Days')
ax2.set_ylabel('expenses')
ax2.grid(color='b', ls = '-.', lw = 0.15) # setting grids

# plot data
ax1.plot(months, savings, c='r')
ax2.plot(days, expenses, c='green')
```

Example:

```python
months = np.array(['Jan', 'Feb', 'March', 'April', 'May', 'Jun'])
savings = np.array([8500, 14700, 12850, 13900, 23650, 15500])
expenses = np.array([1500, 1200, 1000, 500, 2500, 3000, 700])
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

fig, ax = plt.subplots(1, 2, figsize=(10,5), facecolor='c', edgecolor='y', linewidth=5, tight_layout=True)

ax[0].plot(months, savings, c='r') # plot on first axes

ax[0].set_title('Monthly savings', fontsize=15)
ax[0].set_xlabel('Months', fontsize=15)
ax[0].set_ylabel('Savings', fontsize=15)

ax[1].plot(days, expenses, c='g') # plot on second axes
ax[1].set_title('Average Day expenses', fontsize=15)
ax[1].set_xlabel('Days', fontsize=15)
ax[1].set_ylabel('expenses', fontsize=15)
```

Example 2:

```python
X = np.linspace(0,4*np.pi,20)
ypoints1 = np.sin(X)
ypoints2 = np.cos(X)
ypoints3 = np.arccos(ypoints2)
ypoints4 = np.arcsin(ypoints1)

fig, ax = plt.subplots(2, 2, figsize=(10,7), tight_layout=True)

# ax is a 2D array

ax[0][0].plot(X, ypoints1, c='r')
ax[0][0].set_title('sin X')
ax[0][0].set_xlabel('X values')
ax[0][0].set_ylabel('Sin X values')

ax[0][1].plot(X, ypoints2, c='tab:cyan')
ax[0][1].set_title('cos X')
ax[0][1].set_xlabel('X values')
ax[0][1].set_ylabel('cos X values')

ax[1][0].plot(X, ypoints3, c='green')
ax[1][0].set_title('arccos X')
ax[1][0].set_xlabel('X values')
ax[1][0].set_ylabel('arccos ypoints2 values')

ax[1][1].plot(X, ypoints4, 'm')
ax[1][1].set_title('arcsin ypoints1')
ax[1][1].set_xlabel('X values')
ax[1][1].set_ylabel('arcsin ypoints1 values')
```

### Creating subplots with `subplot2grid()` method

```python
X = np.linspace(0,4*np.pi,20)
ypoints1 = np.sin(X)
ypoints2 = np.cos(X)
ypoints3 = np.square(X)
ypoints4 = X**3

fig = plt.figure(figsize=(9, 5), tight_layout=True)
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))

ax1.plot(X, ypoints1, c='r')
ax1.set_title('sin X')
ax1.set_xlabel('X values')
ax1.set_ylabel('Sin X values')

ax2.plot(X, ypoints2, c='tab:cyan')
ax2.set_title('cos X')
ax2.set_xlabel('X values')
ax2.set_ylabel('cos X values')

ax3.plot(X, ypoints3, c='green')
ax3.set_title('X squared')
ax3.set_xlabel('X values')
ax3.set_ylabel('X squared values')

ax4.plot(X, ypoints4, 'm')
ax4.set_title('X cubed')
ax4.set_xlabel('X values')
ax4.set_ylabel('X cubed values')
```

# Setting the X and Y limits and tick labels

## Setting X and Y limits

The X-limits are set with ax.set_xlim(). Syntax:

left is the left xlim in data coordinates.
right is the right xlim in data coordinates
emit is used to notify observers of limit change.
The Y-limits are set with ax.set_ylim(). Syntax:

bottom is the bottom ylim in data coordinates.
top is the top ylim in data coordinates.
emit notifies the observers of limit change.

```python
X = np.linspace(0,4*np.pi,20)
y = np.square(X)

fig, ax = plt.subplots(1, 2, figsize=(8,5), tight_layout=True)

ax[0].plot(y, c='r')
ax[0].set_title('Original limits')
ax[0].set_xlabel('X values')
ax[0].set_ylabel('Sin X values')

ax[1].plot(y, c='g')
ax[1].set_title('Set limits')
ax[1].set_xlabel('X values')
ax[1].set_ylabel('Sin X values')

# setting limits
ax[1].set_ylim(0, 120)
ax[1].set_xlim(0, 10)
```

## Setting tick labels

```python
X = np.linspace(0,4*np.pi,20)
y = np.square(X)

fig, ax = plt.subplots(1, 2, figsize=(8,3), tight_layout=True)

ax[0].plot(y, c='r')
ax[0].set_title('Original')
ax[0].set_xlabel('X values')
ax[0].set_ylabel('Sin X values')

ax[1].plot(y, c='g')
ax[1].set_title('Set X tick labels')
ax[1].set_xlabel('X values')
ax[1].set_ylabel('Sin X values')

# setting limits
ax[1].set_ylim(0, 120)
ax[1].set_xlim(0, 10)
# set labels
ax[1].set_xticks([0, 2, 4, 6, 8, 10])
ax[1].set_xticklabels(['zero','two','four','six', 'eight', 'ten'])
plt.show()
```

# Twin Axes. Sharign same Axes `twinx`

```python
X = np.linspace(0,4*np.pi,20)
y = np.sin(X)
z = np.cos(X)

fig, ax = plt.subplots(figsize=(5, 2), dpi=100)

ax.plot(X, y, c='r')
ax.set_xlabel('X values')
ax.set_ylabel('Sin X values')

ax2 = ax.twinx() # setting the twin axes

ax2.plot(X, z, c='purple')
ax2.set_ylabel('cos X values')

ax.grid(c='g', ls=':')
fig.legend(['sin X', 'cos X'], loc='upper right', bbox_to_anchor=(0.81, 0.87), fontsize=8)
```

# Plot types

## Bar graph

```python
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
expenses = np.array([1500, 1200, 1000, 500, 2500, 3000, 700])

fig, ax = plt.subplots(dpi=100, facecolor=(0.5, 0.5, 0.4, 0.2))

ax.bar(days, expenses, color='m')
ax.set_title('Average Day expenses')
ax.set_xlabel('Days')
ax.set_ylabel('expenses')
```

### Horizontal bar `barh()`

```python
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
expenses = np.array([1500, 1200, 1000, 500, 2500, 3000, 700])

fig, ax = plt.subplots(figsize=(10, 4), facecolor=(0.5, 0.5, 0.4, 0.2))

ax.barh(days, expenses, color='m')
ax.set_title('Average Day expenses')
ax.set_xlabel('Expenses')
ax.set_ylabel('Days')
```

### Stacked Bar graph

```python
x = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
c1 = np.array([6000, 4000, 3500, 2000, 5000])
c2 = np.array([3000, 2500, 4500, 6000, 4500])
c3 = np.array([2500, 6000, 3500, 1500, 3000])
c4 = np.array([7200, 3600, 3400, 5500, 2400])

fig, ax = plt.subplots(figsize=(8, 5), facecolor=(0.5, 0.5, 0.4, 0.2))

ax.bar(x, c1, color='r')
ax.bar(x, c2, bottom=c1, color='g')
ax.bar(x, c3, bottom= c1+c2, color='c')
ax.bar(x, c4, bottom= c1+c2+c3, color='b')

ax.set_title('Day profit from different companies')
ax.set_ylabel('Profis')
ax.set_xlabel('Days')
ax.legend(['Co.1', 'Co.2', 'Co.3', 'Co.4'], loc='best')
```

## Histogram

```python
x = np.array([3000, 400, 500, 6000, 4000, 2500, 6000, 3800, 1500, 3000 ])

fig, ax = plt.subplots()

ax.hist(x, bins=[0, 1500, 2000, 3000, 4000, 5000, 6000]) # made number of bins as a list (6 Bins)
ax.set_title('Histogram')
ax.set_ylabel('Frequency')
ax.set_xlabel('Values')
```

## Pie charts

```python
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
data = np.array([1500, 1200, 1000, 500, 2500, 3000, 700])

fig, ax = plt.subplots(figsize=(10, 4), dpi=140, facecolor=(0.5, 0.5, 0.4, 0.2))

explode = [0.1, 0.5, 0, 0, 0, 0, 0]

ax.pie(data, labels=days, explode=explode, autopct='%1.2f%%', shadow=True)
plt.show()
```

## Scatter Plots

```python
x = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
c1 = np.array([6000, 4000, 3500, 2000, 5000])
c2 = np.array([3000, 2500, 4500, 6000, 4500])
c3 = np.array([2500, 6000, 3500, 1500, 3000])
c4 = np.array([7200, 3600, 3400, 5500, 2400])

fig, ax = plt.subplots(figsize=(8, 5), facecolor=(0.5, 0.5, 0.4, 0.2))

ax.scatter(x, c1, color='r', s=100)
ax.scatter(x, c2, color='g', s=100)
ax.scatter(x, c3, color='c', s=100)
ax.scatter(x, c4, color='b', s=100)

ax.set_title('Day profit from different companies')
ax.set_ylabel('Profis')
ax.set_xlabel('Days')
```
